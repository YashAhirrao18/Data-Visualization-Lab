```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data (student exam scores across subjects)
data = {
    "Math": [85, 90, 78, 92],
    "Science": [88, 85, 91, 89],
    "History": [75, 88, 75, 92],
    "English": [82, 78, 86, 88],
}

# Create a DataFrame from the data
import pandas as pd
df = pd.DataFrame(data)

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df, annot=True, cmap="YlGnBu", linewidths=0.5)
plt.title("Student Exam Scores Heatmap")
plt.show()
```
