```python
from matplotlib import pyplot as plt
import numpy as np
 
 
# Creating dataset
a = np.array([22, 87, 5, 43, 56,
              73, 55, 54, 11,
              20, 51, 5, 79, 31,
              27])
 
# Creating histogram
fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(a, bins = [0, 25, 50, 75, 100])
 
# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
years = [2010, 2011, 2012, 2013, 2014]
categories = ['A', 'B', 'C']
data = {
    'A': [30, 45, 60, 80, 90],
    'B': [10, 20, 30, 40, 50],
    'C': [5, 10, 15, 20, 25]
}

# Create the area plot
plt.stackplot(years, data.values(), labels=data.keys(), alpha=1)
plt.legend(loc='upper left')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Area Plot Example')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data (1000 data points from a normal distribution)
data = np.random.randn(1000)

# Create a histogram
plt.hist(data, bins=30, density=True, alpha=0.7, color='blue')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')

plt.show()
```
